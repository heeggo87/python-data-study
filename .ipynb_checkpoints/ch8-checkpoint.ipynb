```python
# 여러그래프 그리기
# 도화지 생성
fig = plt.figure(figsize=(10,3))

# 그래프 영역 나누기
# 왼쪽 칸 (1행 2열 중 1번째)
ax1 = fig.add_subplot(1,2,1)  
# 오른쪽 칸 (1행 2열 중 2번째)
ax2 = fig.add_subplot(1,2,2)  

# 학생 시험 점수 데이터
students = [1,2,3,4,5] # 학번
scores = [80, 90, 70, 85, 95] # 점수

# 첫번째 그래프 - 선 
ax1.plot(scores)
# 두번째 그래프 - 막대
ax2.bar(students, scores)
```

```python
fit, axs = plt.subplots(1, 2, figsize=(10,3))

# 학생 시험 점수 데이터
students = [1,2,3,4,5] # 학번
scores = [80, 90, 70, 85, 95] # 점수

# 첫번째 그래프 그리기
axs[0].plot(scores)

# 두번째 그래프 그리기
axs[1].bar(students, scores)
```

```python
# 1. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 과일 판매량
# 과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']
# 판매 개수 : [50, 80, 40, 70, 60]

fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)  
ax2 = fig.add_subplot(1,2,2)  

fruit = ['사과', '바나나', '포도', '오렌지', '복숭아']
scores = [50, 80, 40, 70, 60] 

ax1.plot(scores)
ax2.bar(fruit, scores)
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 요일별 유튜브 시청 시간
# 요일 : ['월', '화', '수', '목', '금']
# 시청 시간 : [2, 3, 4, 2, 5]
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)  
ax2 = fig.add_subplot(1,2,2)  

day = ['월', '화', '수', '목', '금']
time = [2, 3, 4, 2, 5] 

ax1.plot(time)
ax2.bar(day, time)
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]

fig, axs = plt.subplots(1, 2, figsize=(10, 3))

day = ['월', '화', '수', '목', '금']
book = [1, 0.5, 2, 1.5, 2.5]

axs[0].bar(day, book)
axs[1].scatter(day, book)
```

```python
# 4. 하나의 도화지에 2개의 그래프(선 그래프, 산점도)를 그리세요.
# 데이터 : 도시별 평균 기온(℃)
# 도시 : ['서울', '부산', '대구', '광주', '제주’]
# 평균 기온 : [15.5, 17.8, 16.2, 15.0, 18.3]

fig, axs = plt.subplots(1, 2, figsize=(10, 3))

city = ['서울', '부산', '대구', '광주', '제주']
weather = [15.5, 17.8, 16.2, 15.0, 18.3]

axs[0].plot(city, weather)
axs[1].scatter(city, weather)
```

```python
df = pd.DataFrame({
    '이름':['영희','철수','민수','지연'],
    '점수':[85,90,80,100],
    '반':['1반','1반','2반','2반']
})
df
```

```python
# 시본으로 여러 그래프 그리기
fig, axs = plt.subplots(1, 2, figsize=(10,3)) # 행열, 캔버스의 크기

sns.barplot(data=df, x='이름', y='점수', ax=axs[0]) # 그래프 위치 설정 axs
axs[0].set_title('학생 점수(막대)')

sns.lineplot(data=df, x='이름', y='점수', ax=axs[1])
axs[1].set_title('학생 점수(선)')

plt.tight_layout() # 겹치지않게 자동으로 간격 조정
```

```python
# 1. 요일별 평균 기온 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 선 그래프
# 두 번째 서브플롯: 산점도

df = pd.DataFrame({
    '요일':['월','화','수','목','금','토','일'],
    '기온':[12,14,13,15,17,18,16]
})
df

fig, axs = plt.subplots(1,2, figsize=(10,3))

sns.lineplot(data=df, x='요일', y='기온', ax=axs[0])
sns.scatterplot(data=df, x='요일', y='기온', ax=axs[1])

axs[0].set_title('요일별 기온(선)')
axs[1].set_title('요일별 기온(산점도)')
```

```python
# 2. 과일별 판매량 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프
# 세 번째 서브플롯: 산점도

df = pd.DataFrame({
    '과일':['사과','바나나','포도','딸기'],
    '판매량':[120,80,140,100]
})
df

fig, axs = plt.subplots(1,3, figsize=(10,3))

sns.barplot(data=df, x='과일', y='판매량', ax=axs[0])
sns.lineplot(data=df, x='과일', y='판매량', ax=axs[1])
sns.scatterplot(data=df, x='과일', y='판매량', ax=axs[2])

axs[0].set_title('과일별 판매량(막대)')
axs[1].set_title('과일별 판매량(선)')
axs[2].set_title('과일별 판매량(산점도)')

plt.tight_layout() # 겹치지않게 자동으로 간격 조정
```

```python
# 3. 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 산점도
# 두 번째 서브플롯: 막대 그래프
# 세 번째 서브플롯: 선 그래프

df = pd.DataFrame({
    '공부시간':[2,3,4,5,6,7,8],
    '점수':[55,60,65,70,75,85,90]
})
df

fig, axs = plt.subplots(3,1, figsize=(6,12))
sns.scatterplot(data=df, x='공부시간', y='점수', ax=axs[0])
sns.barplot(data=df, x='공부시간', y='점수', ax=axs[1])
sns.lineplot(data=df, x='공부시간', y='점수', marker='o', ax=axs[2])

ax=axs[0].set_title('공부시간-점수(산점도)')
ax=axs[1].set_title('공부시간-점수(막대)')
ax=axs[2].set_title('공부시간-점수(선)')

plt.tight_layout()
```

```python
# 4. 과일 판매량과 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯(왼쪽 위)에는 과일 판매량 막대 그래프
# 두 번째 서브플롯(오른쪽 위)에는 과일 판매량 선 그래프
# 세 번째 서브플롯(왼쪽 아래)에는 학생 시험 점수 막대 그래프
# 네 번째 서브플롯(오른쪽 아래)에는 학생 시험 점수 산점도

df_fruit = pd.DataFrame({
    '과일':['사과','바나나','포도','딸기'],
    '판매량':[120,80,140,100]
})

df_score = pd.DataFrame({
    '학생':['영희','철수','민수','지연'],
    '점수':[85,90,80,100],
})

fig, axs = plt.subplots(2,2, figsize=(8,8)) # 2행 2열

sns.barplot(data=df_fruit, x='과일', y='판매량', ax=axs[0,0])
sns.lineplot(data=df_fruit, x='과일', y='판매량', marker='o', ax=axs[0,1])
sns.barplot(data=df_score, x='학생', y='점수', ax=axs[1,0])
sns.scatterplot(data=df_score, x='학생', y='점수', ax=axs[1,1])

axs[0,0].set_title('과일별 판매량(막대)')
axs[0,1].set_title('과일별 판매량(선)')
axs[1,0].set_title('학생 점수(막대)')
axs[1,1].set_title('학생 점수(산점도)')

plt.tight_layout() # 겹치지않게 자동으로 간격 조정
```

```python
df = pd.DataFrame({
    '구': ['서구','서구','연수구','연수구','남동구','남동구'],
    '인구(만명)': [45, 38, 35, 28, 55, 50],
    '동': ['연희동','가좌동','송도동','옥련동','구월동','논현동']
})

# 2. 인천 인구 데이터를 사용해서, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# (1) 막대 그래프
sns.barplot(data=df, x="동", y="인구(만명)", hue="구", palette="pastel", ax=axs[0])
axs[0].set_title("인천 구·동별 인구 현황 (막대)", fontsize=14, fontweight="bold")
axs[0].set_xlabel("동")
axs[0].set_ylabel("인구(만명)")

# 텍스트 추가
axs[0].text("구월동", 55, "55만명", color="red", ha="center", va="bottom")

# (2) 선 그래프 (선스타일 적용)
sns.lineplot(
    data=df, x="동", y="인구(만명)",
    marker="o", linestyle="--", color="blue", ax=axs[1]
)
axs[1].set_title("인천 구·동별 인구 현황 (선)", fontsize=14, fontweight="bold")
axs[1].set_xlabel("동")
axs[1].set_ylabel("인구(만명)")

plt.tight_layout()
```

```python
df = pd.DataFrame({
    '카테고리': ['식품','식품','의류','의류','전자제품','전자제품'],
    '판매량': [120, 150, 80, 95, 200, 220],
    '상품명': ['라면','과자','셔츠','신발','노트북','스마트폰']
})
df

# 4. 상품 판매 데이터를 사용해서, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(data=df, x="상품명", y="판매량", hue="카테고리", ax=axs[0])
axs[0].set_title("상품별 판매량 (막대)", fontsize=14, fontweight="bold")
axs[0].set_xlabel("상품명")
axs[0].set_ylabel("판매량")

sns.lineplot(data=df, x="상품명", y="판매량", marker="s", linestyle=":", ax=axs[1])
axs[1].set_title("상품별 판매량 (선)", fontsize=14, fontweight="bold")
axs[1].set_xlabel("상품명")
axs[1].set_ylabel("판매량")

axs[1].text("스마트폰", 220, "220개", color="red", ha="center", va="bottom")

plt.tight_layout()
```
