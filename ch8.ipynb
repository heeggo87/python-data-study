```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 선 그래프 plot
a = np.arange(1,11)
a
plt.plot(a)
```

```python
# 막대 그래프 bar
a = ['1반', '2반', '3반', '4반']
b = [80, 90, 85, 95]
plt.bar(a, b)
```

```python
# 히스토그램 hist
# 시험 점수 100개 만들기
scores = np.random.randint(50,101,100)
scores

plt.hist(scores, bins=5) 
plt.title('시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')
```

```python
# 여러그래프 그리기
# 도화지 생성
fig = plt.figure(figsize=(10,3))

# 그래프 영역 나누기
# 왼쪽 칸 (1행 2열 중 1번째)
ax1 = fig.add_subplot(1,2,1)  
# 오른쪽 칸 (1행 2열 중 2번째)
ax2 = fig.add_subplot(1,2,2)  

# 학생 시험 점수 데이터
students = [1,2,3,4,5] # 학번
scores = [80, 90, 70, 85, 95] # 점수

# 첫번째 그래프 - 선 
ax1.plot(scores)
# 두번째 그래프 - 막대
ax2.bar(students, scores)
```

```python
fit, axs = plt.subplots(1, 2, figsize=(10,3))

# 학생 시험 점수 데이터
students = [1,2,3,4,5] # 학번
scores = [80, 90, 70, 85, 95] # 점수

# 첫번째 그래프 그리기
axs[0].plot(scores)

# 두번째 그래프 그리기
axs[1].bar(students, scores)
```

```python
# 1. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 과일 판매량
# 과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']
# 판매 개수 : [50, 80, 40, 70, 60]

fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)  
ax2 = fig.add_subplot(1,2,2)  

fruit = ['사과', '바나나', '포도', '오렌지', '복숭아']
scores = [50, 80, 40, 70, 60] 

ax1.plot(scores)
ax2.bar(fruit, scores)
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 요일별 유튜브 시청 시간
# 요일 : ['월', '화', '수', '목', '금']
# 시청 시간 : [2, 3, 4, 2, 5]
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)  
ax2 = fig.add_subplot(1,2,2)  

day = ['월', '화', '수', '목', '금']
time = [2, 3, 4, 2, 5] 

ax1.plot(time)
ax2.bar(day, time)
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]

fig, axs = plt.subplots(1, 2, figsize=(10, 3))

day = ['월', '화', '수', '목', '금']
book = [1, 0.5, 2, 1.5, 2.5]

axs[0].bar(day, book)
axs[1].scatter(day, book)
```

```python
# 4. 하나의 도화지에 2개의 그래프(선 그래프, 산점도)를 그리세요.
# 데이터 : 도시별 평균 기온(℃)
# 도시 : ['서울', '부산', '대구', '광주', '제주’]
# 평균 기온 : [15.5, 17.8, 16.2, 15.0, 18.3]

fig, axs = plt.subplots(1, 2, figsize=(10, 3))

city = ['서울', '부산', '대구', '광주', '제주']
weather = [15.5, 17.8, 16.2, 15.0, 18.3]

axs[0].plot(city, weather)
axs[1].scatter(city, weather)
```
