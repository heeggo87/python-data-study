```python
import numpy as np
import pandas as pd
```

```python
# products.csv 파일을 불러오세요.
df = pd.read_csv('quiz/products.csv')
# print(df)

# 상품명을 인덱스로 설정하세요.
df = df.set_index('상품명')

# 모든 상품명을 출력하세요.
df.index

# 첫 번째 상품(사과)의 가격을 출력하세요.
df.loc['사과','가격']

# 바나나의 재고를 출력하세요.
df.loc['바나나','재고']

# 포도의 가격을 4500으로 수정하세요.
df.loc['포도','가격'] = 4500
df

# 딸기의 재고를 10으로 수정하세요.
df.loc['딸기','재고'] = 10
df
```

```python
# sales.csv 파일을 불러오세요.
df = pd.read_csv('quiz/sales.csv',encoding='euc-kr')
print(df.index)
print(df.columns)
df

# month를 인덱스로 설정하세요.
df = df.set_index('month')
df

# sales 컬럼 전체를 출력하세요.
df['sales']

# 2월의 모든 데이터를 출력하세요.
df.loc['2월']

# 이윤의 합계와 평균을 구하세요. (1400, 350)
df['profit']
print(df['profit'].sum())
print(df['profit'].mean())

# 이윤이 300이상인 행만 출력하세요.
df[df['profit'] >= 300]
```

```python
# weather.csv 파일을 불러오세요.
df = pd.read_csv('quiz/weather.csv',encoding='euc-kr')

df = df.set_index('city')

# 각 도시의 봄 기온(spring)을 모두 출력하세요.
df['spring']

# 서울의 모든 데이터를 출력하세요.
df.loc['서울']

# 겨울(winter)의 평균 기온을 구하세요.
df['winter']
int(df['winter'].mean())


# 여름 기온(summer)이 28도 이상인 도시만 출력하세요
df[df['summer'] >= 28]
```

```python
# products.json 파일을 불러오세요.
df = pd.read_json('quiz/products.json')
# df

# 모든 상품명을 출력하세요.
df = df.set_index('상품명')
df.index

# 가격이 3000원 이상인 상품만 출력하세요.
df[df['가격'] >= 3000]

# 평균 가격을 구하세요.
# 새로운 컬럼 총액(가격*재고)을 추가하세요.
print(df['가격'] * df['재고'])
# 딸기의 가격을 6000으로 수정하세요.
df.loc['딸기'] = 6000
# df
# 재고 컬럼을 삭제하세요.
del df['재고']
df
```

```python
# books.json 파일을 불러오세요.
df = pd.read_json('quiz/books.json')

# 책 제목을 인덱스로 설정하세요.
df = df.set_index('제목')

# 가격이 20000원 이상인 책만 출력하세요.
df [df['가격'] >= 20000]

# 대출가능권수가 2권 이하인 책을 출력하세요.
df[df['대출가능권수'] <= 2]

# 가격이 비싼 순으로 정렬하여 출력하세요.
df.sort_values('가격', ascending=False)

# 대출가능권수가 많은 순으로 정렬하여 출력하세요.
df.sort_values('대출가능권수', ascending=False)

# 전체 평균 가격을 구하세요.
int(df['가격'].mean())
```
